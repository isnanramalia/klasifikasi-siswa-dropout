```python
import pandas as pd

# Membaca kedua file CSV
data_numerik = pd.read_csv("data_numerik.csv")
data_smk_askha_final = pd.read_csv("data_smk_askha_final.csv")

# Fungsi untuk menghasilkan pemetaan otomatis dari nilai unik dalam kolom-kolom kategorikal
def generate_mapping(data):
    mappings = {}
    for column in data.columns:
        if data[column].dtype == 'object':  # Memeriksa kolom dengan tipe data objek (teks)
            unique_values = data[column].unique()
            mapping = {value: idx for idx, value in enumerate(unique_values)}
            mappings[column] = mapping
    return mappings

# Menghasilkan pemetaan untuk data teks ke data numerik
mapping_data_text_to_numeric = generate_mapping(data_smk_askha_final)

# Menampilkan pemetaan yang dihasilkan
for column, mapping in mapping_data_text_to_numeric.items():
    print(f"Pemetaan untuk kolom '{column}': {mapping}")

# Mengubah nilai teks ke nilai numerik dalam data smk askha final
for column, mapping in mapping_data_text_to_numeric.items():
    data_smk_askha_final[column] = data_smk_askha_final[column].map(mapping)

# Menampilkan data smk askha final yang sudah diubah ke nilai numerik
print("\nData SMK Askha Final setelah diubah ke nilai numerik:")
print(data_smk_askha_final)
```

```python
def generate_reverse_mapping(mapping):
    reverse_mappings = {}
    for column, values in mapping.items():
        reverse_mappings[column] = {v: k for k, v in values.items()}
    return reverse_mappings

# Menghasilkan pemetaan dari nilai numerik ke teks
reverse_mapping_data_numeric_to_text = generate_reverse_mapping(mapping_data_text_to_numeric)

# Menampilkan pemetaan dari nilai numerik ke teks
for column, mapping in reverse_mapping_data_numeric_to_text.items():
    print(f"Pemetaan untuk kolom '{column}': {mapping}")

# Mengubah nilai numerik ke nilai teks dalam data smk askha final
for column, mapping in reverse_mapping_data_numeric_to_text.items():
    data_smk_askha_final[column] = data_smk_askha_final[column].map(mapping)

# Menampilkan data smk askha final yang sudah diubah ke nilai teks
print("\nData SMK Askha Final setelah diubah ke nilai teks:")
print(data_smk_askha_final)
```
